```python
#Датасеты проверены на типы данных, пропуски и выбросы
```

```python
df['speed_travel_category']=df['trip_duration']
df.info()
```

```python
#Добавлена дополнительная категориальная переменная по времени поездки
```

```python
df['date'] = df['pickup_datetime'].apply(lambda x: str(x.day) + '-' + str(x.month) + '-' + str(x.year))
df = df.merge(weather, on='date').drop(['date'], axis=1)
df
```

```python
df.info()
```

```python
df['year'] = df['dropoff_datetime'].apply(lambda x: x.year)
df['month'] = df['dropoff_datetime'].apply(lambda x: x.month)
df['day'] = df['dropoff_datetime'].apply(lambda x: x.day)
df['hour'] = df['dropoff_datetime'].apply(lambda x: x.hour)
df
```

```python
q1=df.speed_travel_category.quantile(0.25)
q2=df.speed_travel_category.quantile(0.5)
q3=df.speed_travel_category.quantile(0.75)

df.speed_travel_category=np.where((df.speed_travel_category<=q1),1,df.speed_travel_category)
df.speed_travel_category=np.where(((df.speed_travel_category<q2)&(df.speed_travel_category>q1)),2,df.speed_travel_category)
df.speed_travel_category=np.where(((df.speed_travel_category<q3)&(df.speed_travel_category>=q2)),3,df.speed_travel_category)
df.speed_travel_category=np.where((df.speed_travel_category>=q3),4,df.speed_travel_category)

df.drop('trip_duration', axis=1, inplace=True)
df.info()
```
